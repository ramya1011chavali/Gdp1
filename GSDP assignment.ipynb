```python
c4=gf3['Per Capita GSDP (Rs.)'].quantile(0.2)
c3=gf3['Per Capita GSDP (Rs.)'].quantile(0.5)
c2=gf3['Per Capita GSDP (Rs.)'].quantile(0.85)
c1=gf3['Per Capita GSDP (Rs.)'].quantile(1)
#calculated the Quantile 
```

```python
def cond(gf3):
    if(gf3['Per Capita GSDP (Rs.)']<=c1 and gf3['Per Capita GSDP (Rs.)']>=c2):
        return 'C1'
    elif(gf3['Per Capita GSDP (Rs.)']<=c2 and gf3['Per Capita GSDP (Rs.)']>=c3):
        return 'C2'
    elif(gf3['Per Capita GSDP (Rs.)']<=c3 and gf3['Per Capita GSDP (Rs.)']>=c4):
        return 'C3'
    elif(gf3['Per Capita GSDP (Rs.)']<=c4):
        return 'C4'
    else:
        return 0
gf3['Quan_cat']=gf3.apply(cond,axis=1)
gf3
#Categorized into c1 to c4 with a Function    
    
    
```

```python
c1df=gf3[gf3.Quan_cat=='C1']
c2df=gf3[gf3.Quan_cat=='C2']
c3df=gf3[gf3.Quan_cat=='C3']
c4df=gf3[gf3.Quan_cat=='C4']
#categoried the values
```

```python
#grouping the Category
cat1=gf3.groupby(['Quan_cat']).sum()
cat1
```

```python
cat1.drop(columns=['Primary','Crops','Livestock','Forestry and logging','Fishing and aquaculture','Trade & repair services','Hotels & restaurants','Railways','Road transport','Water transport','Air transport','Services incidental to transport','Storage','Communication & services related to broadcasting','Secondary','Tertiary',
       'TOTAL GSVA at basic prices', 'Taxes on Products',
       'Subsidies on products'],axis=1,inplace=True)
```

```python
cat1.drop(columns=['Per Capita GSDP (Rs.)','Services incidental to transport*','Trade & repair services*','Road transport**','Road transport*'],axis=1,inplace=True) 
```

```python
cat1.drop(columns=['State'],axis=1,inplace=True)
cat1
#this error is because i have this twice and it showing the error as it is already dropped 
```

```python
catli=[]
for ct in list(cat1.index):
    cat2=cat1.T[ct]
    cat2=cat2.reset_index()
    cat2.rename(columns={ct:'Values','index':'Items'},inplace=True)
    cat2=cat2.sort_values(by='Values',ascending=False)
    cat2=cat2.reset_index(drop=True)
    category_gsdp_val=list(cat2['Values'])[0]
    def get_contri_percent(x):
        ret_val=round(100*(x/category_gsdp_val),2)
        return ret_val
    cat2['%Contri in GDP']=cat2['Values'].apply(get_contri_percent)
    end=4
    while cat2.loc[1:end,'%Contri in GDP'].sum()<79:
        end=end+1
    cat2=cat2.loc[1:end]
    catli.append(cat2)
#calculated the subsectors percentage upto 80%
```

```python
#plotting contri to sectors having approx.80% contri to gsdp to different states categories
figure=plt.figure(figsize=(10,15))
for index in range (len(catli)):
    plt.subplot(4,1,index+1)
    ax=sns.barplot(x=catli[index]['%Contri in GDP'],y=catli[index]['Items'],data=catli[index])
    plt.xlabel('')
    plt.ylabel('ct:'+str(index+1),fontsize=14)
    plt.yticks(fontsize=12)
    plt.xticks(np.arange(0,25,3),fontsize=10)
    plt.xlim(0,25)
plt.xlabel('%Contribution of items in GSDP',fontsize=16)
plt.show()
```
